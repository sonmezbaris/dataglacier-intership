# Cab transaction id city and customer id data merged based on city

```python
master_data=pd.merge(cab_transaction_customer_data, city, how="left", on="City")

master_data
```

```python
master_data["Date of Travel"].describe()
```

# Master data info

```python
master_data.info()
```

```python
master_data["Population"]=[x.replace(",","") for x in master_data["Population"]]
master_data["Users"]=[x.replace(",","") for x in master_data["Users"]]

## changing the data types

master_data["Population"]=master_data["Population"].astype(float)

master_data["Users"]=master_data["Users"].astype(float)
```

## Change Format master data date of travel column Year-Month-Day

```python
a = master_data['Date of Travel']
## using the base date as follow
base_date = pd.Timestamp('1900-01-01')
dates = [base_date + pd.DateOffset(date_offset) for date_offset in a]
master_data['date_of_travel'] = pd.to_datetime(dates, format='%y/%m/%d')
```

```python
master_data['date_of_travel'].value_counts()
```

```python
master_data
```

### Number of cab companies

```python
master_data["year"]=master_data["date_of_travel"].dt.year

master_data["month"]=master_data["date_of_travel"].dt.month
master_data["day"]=master_data["date_of_travel"].dt.day
master_data["week"]=master_data["date_of_travel"].dt.isocalendar().week 
```

```python
master_data['Profit']=master_data['Price Charged']-master_data['Cost of Trip']

master_data
```
